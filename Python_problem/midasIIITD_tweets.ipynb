### Importing useful libraries 

```python
import tweepy 
import json
import jsonlines
import pandas as pd
import datetime
```

### keys to fetch data using Twitter API

```python
consumer_key = "" 
consumer_secret = ""
access_key = ""
access_secret = ""
```

#### Authentication and API call 

```python
# Authorization to consumer key and consumer secret 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
  
# Access to user's access key and access secret 
auth.set_access_token(access_key, access_secret) 
  
# Calling api 
api = tweepy.API(auth)

        
```

#### This code cell fetches all the information of each tweet of the user midasIIITD

```python
whole_info = dict()
tweet_count = 0
for status in tweepy.Cursor(api.user_timeline, id="midasIIITD").items():
    tweet_count += 1
    whole_info[tweet_count] = status._json 
```

```python
# number of tweets by midasIIITD
len(whole_info)
```

#### the cell below first dumps the information extracted into a json and then loads it into a python object known as  whole_info and then writes that python object into a json lines file known as twitter_username.jsonl

```python
whole_info = json.loads(json.dumps(whole_info))
with jsonlines.open('twitter_username.jsonl', 'w') as outfile:
    outfile.write(whole_info)
```

#### the following cell dumps all the information into a python list, this is done to extract specific information of tweets  

```python
info = list()
with jsonlines.open('twitter_username.jsonl', 'r') as readfile:
    for tweet in readfile:
        info.append(tweet)
info[0]     
```

```python
type(info[0])
```

```python
# assigning the info element to a variable name username_whole_twitter_info
username_whole_twitter_info = info[0]
```

```python
# confirming the number of tweets
len(username_whole_twitter_info)
```

#### the block below is used to spilt date and time and to convert the time into 12 hr format and returns date and time of a tweet

```python
def date_time_split_convert(tweet_no):
    date_time_str = username_whole_twitter_info[str(tweet_no+1)]['created_at']
    date = date_time_str[:10] + ' ' + date_time_str[26:30]
    d = datetime.datetime.strptime(date_time_str[11:16], "%H:%M")
    time = d.strftime("%I:%M %p")
    return date, time
```

#### the following block fetches specific information such as date, time, favorite count, etc and creates a list of dictionaries, with each dictionary representing information of one tweet

```python
tweets_list = list()
for i in range(len(username_whole_twitter_info)):
    tweet_info = dict()
    # calling date_time_split_convert() to get date and time
    tweet_info['Date'], tweet_info['Time'] = date_time_split_convert(i)
    tweet_info["Text"] = username_whole_twitter_info[str(i+1)]['text']
    tweet_info["Retweets"] = username_whole_twitter_info[str(i+1)]['retweet_count']
    tweet_info["Likes"] = username_whole_twitter_info[str(i+1)]['favorite_count']
    try:
        media = username_whole_twitter_info[str(i + 1)]['entities']['media']
        tweet_info['No._of_Images'] = len(media)
    except:
        tweet_info['No._of_Images'] = 'None'
    
    tweets_list.append(tweet_info)
```

#### the cell below creates a dataframe of the list created above

```python
df = pd.DataFrame(tweets_list)
```
